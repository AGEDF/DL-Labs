```python
irrelevant_columns = ["GENDER","PREGNANT","COPD","ASTHMA","INMSUPR","OTHER_DISEASE","CARDIOVASCULAR",
"OBESITY","TOBACCO"]
```

```python
df.drop(columns=irrelevant_columns,inplace=True)
```

```python
df.head()
```

```python
df = pd.get_dummies(df,columns=["MEDICAL_UNIT","CLASIFFICATION_FINAL"],drop_first=True)
```

```python
from sklearn.preprocessing import RobustScaler
scaler = RobustScaler()
df.AGE = scaler.fit_transform(df.AGE.values.reshape(-1,1))
```

```python
x = df.drop(columns="DEATH")
y = df["DEATH"]
from sklearn.model_selection import train_test_split
train_x, test_x, train_y, test_y = train_test_split(x,y, test_size=0.4, random_state=42)
print("Train_x :",train_x.shape)
print("Test_x :",test_x.shape)
print("Train_y :",train_y.shape)
print("Test_y :",test_y.shape)
```

```python
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()
logreg.fit(train_x,train_y)
print("Logistic Regression Accuracy :",logreg.score(test_x, test_y))
```

```python
from sklearn.metrics import f1_score
print("Logistic Regression F1 Score :",f1_score(test_y,logreg.predict(test_x),average=None))
```

```python
from sklearn.metrics import confusion_matrix
sns.heatmap(confusion_matrix(test_y, logreg.predict(test_x)), annot=True, fmt=".0f")
plt.title("Logistic Regression Confusion Matrix",fontsize=18, color="red");
```

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import make_classification
from sklearn.metrics import accuracy_score
base_estimator = DecisionTreeClassifier(max_depth=3)
adaboost_clf = AdaBoostClassifier(base_estimator=base_estimator, n_estimators=50, random_state=42)
adaboost_clf.fit(train_x, train_y)
```

```python
y_pred = adaboost_clf.predict(test_x)
accuracy = accuracy_score(test_y, y_pred)
print("Accuracy:", accuracy)
Accuracy: 0.8796003659500146
print("AdaBoost F1 Score :",f1_score(test_y,adaboost_clf.predict(test_x),average=None))
```

```python
sns.heatmap(confusion_matrix(test_y, adaboost_clf.predict(test_x)), annot=True, fmt=".0f")
plt.title("AdaBoost Confusion Matrix",fontsize=18, color="black");
```
