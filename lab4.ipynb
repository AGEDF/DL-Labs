```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))

x = np.linspace(-10,10)
plt.plot(x,sigmoid(x))
plt.axis('tight')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Sigmoid Activation Function')
plt.show()
```

```python
def linear(x):
    return x

x = np.linspace(-10,10)
plt.plot(x,linear(x))
plt.axis('tight')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Linear Activation Function')
plt.show()
```

```python
def tanh(x):
    return (np.exp(x) - np.exp(-x))/(np.exp(x) + np.exp(-x))
x = np.linspace(-10,10)
plt.plot(x,tanh(x))
plt.axis('tight')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Tanh Activation Function')
plt.show()
```

```python
def Relu(x):
    return [max(0,val) for val in x]

x = np.linspace(-10,10)
plt.plot(x,Relu(x))
plt.axis('tight')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Relu Activation Function')
plt.show()
```

```python
def LeakyRelu(x):
    return [max(0.1*val,val) for val in x]

x = np.linspace(-10,10)
plt.plot(x,LeakyRelu(x))
plt.axis('tight')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Leaky Relu Activation Function')
plt.show()
```

```python
def step(x):
    return [1 if val>0 else 0 for val in x]

x = np.linspace(-10,10)
plt.plot(x,step(x))
plt.axis('tight')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Step Activation Function')
plt.show()
```

```python
def softmax(x):
    exps = np.exp(x - np.max(x, axis=-1))
    return exps / np.sum(exps, axis=-1)

x = np.linspace(-10,10)
plt.plot(x,softmax(x))
plt.axis('tight')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Softmax Activation Function')
plt.show()
```
